```python
import numpy as np
import pandas as pd
import sklearn
 
# Read dataset
df = pd.read_csv("Police_Use_of_Force_preprocessed.csv")
df.head()
print(df.shape)
```

```python
problems = ['Assault in Progress', 'Disturbance', 'Domestic Abuse-In Progress', 'Emotionally Disturb Person', 'Fight', 'Suspicious Person', 'Suspicious Vehicle', 'Traffic Law Enforcement', 'Unknown Trouble', 'Unwanted Person']
prob = []
for i in range(len(df['Problem_m'])):
  if df['Problem_m'][i] in problems:
    prob.append(problems.index(df['Problem_m'][i]))
  else:
    prob.append(10)
print(prob)
prob = pd.DataFrame(prob, columns=['Problem'])
```

```python
is911 = []
for i in range(len(df['Problem_m'])):
  if df['Is911Call'][i] == 'Yes':
    is911.append(1)
  else:
    is911.append(0)
print(is911)
is911 = pd.DataFrame(is911, columns=['Is911Call'])
```

```python
offenses = ['ASLT4', 'DISCON', 'FORCE', 'NARC', 'OBSTRU']
offense = []
for i in range(len(df['Problem_m'])):
  if df['PrimaryOffense'][i] in offenses:
    offense.append(offenses.index(df['PrimaryOffense'][i]))
  else:
    offense.append(5)
print(offense)
offense = pd.DataFrame(offense, columns=['PrimaryOffense'])
```

```python
injury = []
for i in range(len(df['Problem_m'])):
  if df['SubjectInjury'][i] == 'Yes':
    injury.append(1)
  elif df['SubjectInjury'][i] == 'No':
    injury.append(0)
  else:
    injury.append(2)
print(injury)
injury = pd.DataFrame(injury, columns=['SubjectInjury'])
```

```python
roles = ['A', 'ARRESTEE', 'OT', 'S', 'SUSPECT']
role = []
for i in range(len(df['Problem_m'])):
  if df['SubjectRole'][i] in roles:
    role.append(roles.index(df['SubjectRole'][i]))
  else:
    role.append(5)
print(role)
role = pd.DataFrame(role, columns=['SubjectRole'])
```

```python
forces = ["Bodily Force", "Chemical Irritant", "Taser", "Police K9 Bite", "Firearm", "Baton", "Improvised Weapon", 
    "Gun Point Display", "Maximal Restraint Technique", "Less Lethal", "Less Lethal Projectile"]
force = []
for i in range(len(df['Problem_m'])):
    force.append(forces.index(df['ForceType'][i]))
print(force)
# force = pd.DataFrame(force, columns=['ForceType'])
```

```python
races = ['Black','Asian','White','Native American','Other / Mixed Race','Unknown','Pacific Islander']
race = []
for i in range(len(df['Problem_m'])):
    race.append(races.index(df['Race'][i]))
print(race)
race = pd.DataFrame(race, columns=['Race'])
```

```python
sexes = ['Male','Female','Unknown']
sex = []
for i in range(len(df['Problem_m'])):
    sex.append(sexes.index(df['Sex'][i]))
print(sex)
sex = pd.DataFrame(sex, columns=['Sex'])
```

```python
resistances = ['Fled on Foot','Commission of Crime','Verbal Non-Compliance','Tensed','Assault Officer','Fled in Vehicle','Unspecified','Assault Police Horse','Other']
res = []
for i in range(len(df['Problem_m'])):
  if df['TypeOfResistance'][i] in resistances:
    res.append(resistances.index(df['TypeOfResistance'][i]))
  else:
    res.append(9)
print(res)
res = pd.DataFrame(res, columns=['TypeOfResistance'])
```

```python
neighborhoods = ['Hawthorne','Downtown West','Other','Jordan','Near - North','Willard - Hay']
neighborhood = []
for i in range(len(df['Problem_m'])):
    neighborhood.append(neighborhoods.index(df['Neighborhood_m'][i]))
print(neighborhood)
neighborhood = pd.DataFrame(neighborhood, columns=['Neighborhood'])
```

```python

f = [col for col in df if col != "ForceType" and col != "X" and col != "Y" and col != "PoliceUseOfForceID" 
            and col != "CaseNumber" and col != "ResponseDate"  and col != "DateAdded"  and col != "OBJECTID"
            and col != "ForceTypeAction_m" and col != "Neighborhood_m"
            and col != "Neighborhood"
            and col != "ForceTypeAction" and col != "Problem"
            and col != "Problem_m" and col != "Sex"
            and col != "TypeOfResistance" and col != "TypeOfResistance_m" 
            and col != "SubjectRole" and col != "SubjectRole_m" 
            and col != "PrimaryOffense_m" and col != "PrimaryOffense"
            and col != "Is911Call" and col != "Race"
            and col != "SubjectInjury" and col != "SubjectInjury_m" ]
print(f)
data = df[f]
data.head()
data.shape
```

```python
print(data.shape)
data = pd.concat([data, prob, is911, offense, injury, role, race, sex, res, neighborhood], axis=1)
data.head()
```

```python
from interpret.glassbox import ExplainableBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

d = data.shape[1]
n = len(data["EventAge"])
print(n)
labels_string = ["EventAge"]
y = force
vals_from_string = data.filter(items=labels_string).applymap(lambda x: float(x) if not np.isnan(x) else 0.0)
print(vals_from_string.isnull().values.any())
x_no_string = pd.concat([data.iloc[:,1:4], vals_from_string, data.iloc[:, 5:d]], axis=1)
Xoffset =  pd.concat([x_no_string, pd.DataFrame(np.ones(n))], axis=1)

X_train, X_test, y_train, y_test = train_test_split(Xoffset, y, test_size=0.2, random_state=1)
num_labels = Counter(y_train)
print(y)
over = SMOTE(sampling_strategy={1:num_labels[1]+480, 2:num_labels[2]+480, 3:num_labels[3]+480, 4:num_labels[4]+480, 5:num_labels[5]+480, 6:num_labels[6]+480, 7:num_labels[7]+480, 8:num_labels[8]+480, 9:num_labels[9]+480, 10:num_labels[10]+480})
under = RandomUnderSampler(sampling_strategy={0:num_labels[0]-100})
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)
X_SMOTE, y_SMOTE = pipeline.fit_resample(X_train, y_train)
```

```python
ebm = ExplainableBoostingClassifier()
ebm.fit(X_SMOTE, y_SMOTE)
```

```python
from interpret import show

ebm_global = ebm.explain_global(name='EBM')
show(ebm_global)
```

```python
def misclassification(pred, y):
    sum = 0
    for i in range(len(pred)):
        if pred[i] != y[i]:
            sum +=1
    return sum/len(pred)

pred = ebm.predict(X_test)
print(misclassification(pred, y_test))
```
